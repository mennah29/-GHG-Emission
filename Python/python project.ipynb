```python
import plotly.express as px
import pandas as pd

# لوحة الألوان
custom_colors = ["#053B50", "#176B87", "#008891", "#64CCC5", "#98DED9", "#D1F8EF", "#FFFFFF"]

# نجمع القيم على مستوى income_group و sector
df_top_ten = df.groupby(["income_group", "sector"], as_index=False)["emissions_ktco2e"].sum()

# نحسب النسبة المئوية
df_top_ten["percentage"] = df_top_ten["emissions_ktco2e"] / df_top_ten.groupby("income_group")["emissions_ktco2e"].transform("sum") * 100

# نرسم stacked bar مع النسبة في النص
fig = px.bar(
    df_top_ten,
    x="income_group",
    y="emissions_ktco2e",
    color="sector",
    title="Emissions by Income Group and Sector (with Percentages)",
    width=1000,
    height=600,
    color_discrete_sequence=custom_colors,
    text=df_top_ten["percentage"].apply(lambda x: f"{x:.1f}%")
)

# تعديل النص ليكون في منتصف كل جزء
fig.update_traces(
    textposition="inside",
    insidetextanchor="middle",
    textfont_size=12,
)

fig.update_layout(
    xaxis_title="Income Group",
    yaxis_title="Emissions (ktCO2e)",
    legend_title="Sector",
    barmode="stack"
)
fig.update_layout(
    title_font_size=20,
       title_font_weight=700,
    
    xaxis_title_font_size=12,
    xaxis_title_font_weight=500,# عنوان المحور X
    yaxis_title_font_size=12, 
    yaxis_title_font_weight=500,# عنوان المحور Y
    legend_font_size=10          # خط الـ Legend
)

fig.show()
```

```python
# Suppose your DataFrame is named df
# First, create a new column for decade grouping
df['decade'] = pd.cut(df['year'],
                      bins=[1990, 2000, 2010, 2020, 2025],
                      labels=['1990-1999', '2000-2009', '2010-2019', '2020-2024'],
                      right=False)
 
# Now group by decade and sum emissions
emissions_by_decade = df.groupby('decade')['emissions_ktco2e'].sum().reset_index()
 
print(emissions_by_decade)
```

```python
# assuming emissions_by_decade is calculated already
# emissions_by_decade contains two columns: 'decade' and 'emissions_ktco2e'
 
plt.figure(figsize=(10,6))
plt.plot(emissions_by_decade['decade'], emissions_by_decade['emissions_ktco2e'], marker='o',color="#64CCC5")
 
plt.title('Total Emissions by Decade (1990-2024)', fontsize=20, color="#27374D",fontweight=700)
plt.xlabel('Decade', fontsize=12, color="#176B87")

plt.ylabel('Emissions (kt CO2e)',fontsize=12, color="#176B87")
plt.grid(True)
plt.show()
```

```python
# 1. Filter LULUCF sector and negative emissions
lulucf_negative = df[(df['sector'] == 'LULUCF') & (df['emissions_ktco2e'] < 0)]
 
# 2. Convert emissions to positive values
lulucf_negative['emissions_positive'] = lulucf_negative['emissions_ktco2e'].abs()
 
# 3. Sum emissions by country and get top 5
top5_countries = lulucf_negative.groupby('country')['emissions_positive'].sum().nlargest(5).reset_index()
 
# 4. Bar chart
plt.figure(figsize=(10,6))
plt.bar(top5_countries['country'], top5_countries['emissions_positive'],color="#64CCC5")
 
plt.title('Top 5 Countries (LULUCF - Converted Positive Emissions)', fontsize=20, color="#27374D",fontweight=700)
plt.xlabel('Country', fontsize=12, color="#176B87")

plt.ylabel('Emissions (Positive kt CO2e)',fontsize=12, color="#176B87")
plt.xticks(rotation=45)
plt.show()
```

```python
population_df = pd.read_excel(r"C:\Users\HP\Downloads\emission project\populationn.xlsx")   
population_df
```

```python
merged_ddff = pd.merge(emissions_2010, population_df, on='country', how='inner')
merged_ddff
```

```python
merged_ddff['emissions_per_capita'] = merged_ddff['emissions_ktco2e'] / merged_ddff['pop_2010_millions']
merged_ddff
```

```python
dp_median = merged_ddff['gdp_pc_2010_usd'].median()
pop_median = merged_ddff['pop_2010_millions'].median()
emissions_total_median = merged_ddff['emissions_ktco2e'].median()
emissions_percap_median = merged_ddff['emissions_per_capita'].median()
```

```python
high_pop = merged_ddff[merged_ddff['pop_2010_millions'] >= pop_median]
low_pop = merged_ddff[merged_ddff['pop_2010_millions'] < pop_median]

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.barh(high_pop.sort_values('emissions_per_capita', ascending=False).head(10)['country'],
         high_pop.sort_values('emissions_per_capita', ascending=False).head(10)['emissions_per_capita'], color="#053B50")
plt.title('High Population: Top Emissions per Capita',fontsize=14,fontweight=700, color="#27374D")
plt.xlabel('Emissions per Capita',fontsize=12, color="#176B87")

plt.subplot(1,2,2)
plt.barh(low_pop.sort_values('emissions_per_capita', ascending=False).head(10)['country'],
         low_pop.sort_values('emissions_per_capita', ascending=False).head(10)['emissions_per_capita'], color="#64CCC5")
plt.title('Low Population: Top Emissions per Capita',fontsize=14,fontweight=700, color="#27374D")
plt.xlabel('Emissions per Capita',fontsize=12, color="#176B87")
plt.tight_layout()
plt.show()
```

```python
op_median = merged_ddff['pop_2010_millions'].median()
emissions_median = merged_ddff['emissions_ktco2e'].median()

low_pop_high_em = merged_ddff[(merged_ddff['pop_2010_millions'] < pop_median) & 
                             (merged_ddff['emissions_ktco2e'] > emissions_median)]


top5_low_pop_high_em = low_pop_high_em.sort_values('emissions_ktco2e', ascending=False).head(5)

plt.figure(figsize=(8,5))
plt.barh(top5_low_pop_high_em['country'], top5_low_pop_high_em['emissions_ktco2e'], color="#64CCC5")
plt.title('Low Population, High Total Emissions (Top 5)', color="#27374D",fontweight=700,fontsize=20)
plt.xlabel('Total Emissions (ktCO2e)',fontsize=12, color="#176B87")
plt.ylabel('Low_population_contries',fontsize=12, color="#176B87")
plt.show()
```
